## 1. Load and standardize data

```python
from sklearn.datasets import load_iris, load_boston
from sklearn import preprocessing
import pandas as pd
```

```python
data = load_iris()
```

```python
X = data.data
# X
```

```python
# Standardizing the scores
X = preprocessing.scale(X)
# X
```

## 2. Create clusters

```python
from sklearn.cluster import KMeans
```

```python
kmeans_cluster = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
kmeans_cluster
```

```python
kmeans_cluster.labels_
```

```python
# 4 feature values per cluster?
kmeans_cluster.cluster_centers_
```

## 3. visualization

```python
import matplotlib.pyplot as plt
```

```python
# Or we can just do below to get at the labels
kmeans_cluster_index = KMeans(n_clusters=3, random_state=0).fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans_cluster_index)
```

```python
# Adding centers
plt.scatter(X[:, 0], X[:, 1], c=kmeans_cluster_index)

centers = kmeans_cluster.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red')
```
