## 4. Represent each action sequence by n-grams (n=2)

```python
import nltk
from nltk import ngrams
# from nltk.tokenize import word_tokenize
```

```python
# Reshaped the actions into wide form per person
hw5datas2 = hw5data.pivot(index='TestTakerID', columns='ActSeq', values=['Coding'])
```

```python
hw5datas2.head(20)
```

```python
# addled a columns that combined all actions.
hw5datas2['ActionCombined']= hw5datas2.values.tolist()
```

```python
hw5datas2
```

```python
# Added a columns at the end with a list of bigrams.
hw5datas2['bigrams'] = hw5datas2['ActionCombined'].apply(lambda row: list(nltk.ngrams(row, 2)))
```
