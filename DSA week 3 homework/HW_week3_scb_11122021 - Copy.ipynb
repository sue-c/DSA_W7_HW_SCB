### Week 3 homework 2 (Parcing XML file)

```python
# load package
import xml.etree.ElementTree as et  # package for xml parsing
```

```python
# Specify XML file name

hw3xml = r'C:/ETSProjects_SCB/UpskillTraining_2022/Week3/Data and Homework/week3_homework.xml'
```

```python
tree = et.parse(hw3xml)
root = tree.getroot()
```

```python
print(root)

# check how many child 
print(len(root)) # len(object) finds the length of string?
print(len(root[0]))
```

```python
# find out the child of the root and the number of grandchild
for chd in root[0]:
    print(chd.tag,',',chd.text, ',',len(chd))
```

```python
# check the child of the of the first entry of 'event sequence' 
for chd in root[0][5][0]:
    print(chd.tag,',',chd.text, ',',len(chd))
```

```python
hw3xml_df_cols = ["eventName", "eventStartTime", "eventEndTime", "eventBy", "eventTo", "eventResult"]
rows = []

for node in root[0][5]: 
    s_eventName = node.find("eventName").text if node is not None else None
    s_eventStartTime = node.find("eventStartTime").text if node is not None else None
    s_eventEndTime = node.find("eventEndTime").text if node is not None else None
    s_eventBy = node.find("eventBy").text if node is not None else None
    s_eventTo = node.find("eventTo").text if node is not None else None
    s_eventResult = node.find("eventResult").text if node is not None else None
    
    rows.append({"eventName": s_eventName, "eventStartTime": s_eventStartTime, "eventEndTime": s_eventEndTime, "eventBy": s_eventBy, "eventTo": s_eventTo, "eventResult": s_eventResult})
    
hw3xml_df = pd.DataFrame(rows, columns = hw3xml_df_cols)
```
